```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

query = "(from:elonmusk) until:2022-12-31 since:2022-12-01"
tweets = []
limit = 1000

for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    if len(tweets) == limit:
        break
    else:
        tweets.append([tweet.date, tweet.username, tweet.content, tweet.outlinks])
        
df_tweets = pd.DataFrame(tweets, columns=['Date', 'User', 'Tweet', 'Outlinks'])
```

```python
df_tweets.to_parquet('df.parquet.gzip', compression='gzip')  
df_tweets = pd.read_parquet('df.parquet.gzip')  
```
